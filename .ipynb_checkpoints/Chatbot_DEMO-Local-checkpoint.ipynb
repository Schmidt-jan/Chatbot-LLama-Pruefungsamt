```python
import os

# Add a folder to Colab called "Knowledge Base" and upload PDF files of your choice into it.
DOCS_PATH = "main_data/"

# retrieve all PDF files in said folder
files = os.listdir(DOCS_PATH)
pdf_files = [os.path.join(DOCS_PATH, file) for file in files if file.endswith(".pdf")]
```

```python
from haystack.nodes import PDFToTextConverter

converter_pdf = PDFToTextConverter(remove_numeric_tables=True, valid_languages=["de"])

doc_pdf_all = [converter_pdf.convert(file_path=path, meta=None)[0] for path in pdf_files]
```

```python
from haystack.nodes import PreProcessor

# This is a default usage of the PreProcessor.
# The split_length is essential in managing document chunks.
preprocessor = PreProcessor(
    clean_empty_lines=True,
    clean_whitespace=True,
    clean_header_footer=True,
    split_by="word",
    split_length=500,
    split_respect_sentence_boundary=True,
)
docs_default = preprocessor.process(doc_pdf_all)
```

```python
from haystack.document_stores import InMemoryDocumentStore

# Setup Document Store
document_store = InMemoryDocumentStore(use_bm25=True)
document_store.write_documents(docs_default)
```

```python
from haystack.nodes import BM25Retriever

# Initialize Retriever
bm25_retriever = BM25Retriever(document_store=document_store)
```

```python
def compose_query(reader_k, docs, input_text):
    query = f"""
    Answer the question truthfully based solely on the given documents.
    Answer by grouping the given documents into {reader_k} bullet points.
    If they are contextually overlapping, reduce the amount of bulletpoints as you see fit.
    Documents:
    {str(docs)}

    Question: {input_text}
    """
    return query
```

```python
# TEST ENVIRONMENT
input_text = query
reader_k = 3

default_init = "What can I do for you?"
messages=[
        {"role": "system", "content": default_init},
]
docs = p_extractive_premade.run(
        query=input_text, params={"Retriever": {"top_k": reader_k}}
)

# This is the prompt template from slide 26 of the presentation
query = compose_query(reader_k, docs, input_text)

messages.append({"role": "user", "content": query})

```

```python
# Send request to local webserver running the model
response = openai.ChatCompletion.create(
  model="llama-2-13b-chat.Q4_K_M.gguf",
  messages=messages,
  stream=True
)


# Stream response of local llm
reply = ""
for token in response:
    #print(token.choices[0])
    
    item = token.choices[0]["delta"].get("content", "")
    reply += item
    print(item, sep=' ', end='', flush=True)
    
```

# Optional Chatbot Interface from gradio

```python
import gradio as gr

def chatbot(input_text):
    reader_k = 3

    default_init = "What can I do for you?"
    messages=[
            {"role": "system", "content": default_init},
    ]
    docs = p_extractive_premade.run(
        query=input_text, params={"Retriever": {"top_k": reader_k}}
    )

    query = f"""
    Answer the question truthfully based solely on the given documents.
    If a subset of documents are contextually overlapping, reduce the amount of bulletpoints by only keeping one of those.
    If a given document doesn't serve the question, don't respond with its contents.
    Answer by grouping the given documents into {reader_k} bullet points.
    Documents:
    {str(docs)}

    Question: {input_text}
    """
    messages.append({"role": "user", "content": query})
    try:
      response = openai.ChatCompletion.create(
        model="llama-2-13b-chat.Q4_K_M.gguf",
        messages=messages
      )
    except Exception as e:
        return e
    return response.choices[0]['message']['content']

iface = gr.Interface(fn=chatbot,
                    inputs = gr.Textbox(
                        label="Input",
                        info="Ask me questions about the EU AI Act!",
                        lines=3
                    ),
                    outputs="text",
                    title="Chatbot Demo"
                    )

iface.launch(share=True, debug=True)
```
