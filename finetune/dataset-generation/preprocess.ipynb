```python
import fitz
import glob

def extract_paragraphs(text):
    paragraphs = text.split('\n\n')  # Split text into paragraphs based on double newlines (approximately)
    return paragraphs

# works relatively well for html documents aswell
def process_pdf(document):
    all_paragraphs = []

    for page_num in range(document.page_count):
        page = document.load_page(page_num)
        text = page.get_text("text")
        
        paragraphs = extract_paragraphs(text)
        all_paragraphs.extend(paragraphs)

        #print(f"Page {page_num + 1}:")
        for para in paragraphs:
            print(para)
            print()

    return all_paragraphs

def paragraphs_to_json(paragraphs):
    json = []
    for paragraph in paragraphs:
        json.append({
            'paragraph': paragraph
        })
    return json

downloads = glob.glob("downloads/*")

data = []

for download in sorted(downloads):
    #print(f"Processing File: {download}")
    document = fitz.open(download)
    paragraphs = process_pdf(document)
    #print(f"NUM EXTRACTED PARAGRAPHS: {len(paragraphs)}")
    #for paragraph in paragraphs:
    #    print(len(paragraph))
    # of a single document
    paragraphs_json = paragraphs_to_json(paragraphs)
    data.extend(paragraphs_json)

print("\n\n\n")
print(len(data))
```

```python
# Split into test and validation set
from sklearn.model_selection import train_test_split
import os 
import json

finetune_topic = "chatbot-documents"

output_path_train = f"datasets/{finetune_topic}/train.json"
output_path_validation = f"datasets/{finetune_topic}/validation.json"

train_data, validation_data = train_test_split(data, train_size=0.8, test_size=0.2, shuffle=False)
print(f"LENGTH TRAINING: {len(train_data)}")
print(f"LENGTH VALIDATION: {len(validation_data)}")
print(f"LENGTH TOTAL: {len(data)}")

os.makedirs(os.path.dirname(output_path_train), exist_ok=True)
with open(output_path_train, "w", encoding="utf-8") as output_file:
    json.dump(train_data, output_file, indent=2)

os.makedirs(os.path.dirname(output_path_validation), exist_ok=True)
with open(output_path_validation, "w", encoding="utf-8") as output_file:
    json.dump(validation_data, output_file, indent=2)
```
